```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

#### 绘制图形

```python
plt.plot(x,y)
plt.show()
```

#### 绘制多条曲线

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy, color='red', linestyle='--')
plt.show()
```

```python
plt.plot(x,siny,label=('sin(x)'))
plt.plot(x,cosy, color='red', linestyle='--',label=('cos(x)'))
plt.axis([-1,11,-2,2]) # == plt.xlim(-5,15) plt.ylim(0,1.5)
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.title('Welcome to the ML World!')
plt.show()
```

### 绘制散点图

```python
plt.scatter(x,siny)
plt.scatter(x,cosy)
plt.show()
```

```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)

plt.scatter(x,y,alpha=0.5)  # alpha 透明度
plt.show()
```
